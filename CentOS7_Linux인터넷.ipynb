### 1.계정에 관한 에러

CentOS에서 yum 명령어로 install할때 아래의 에러가 발생

```
loaded plugins: fastesmirror, langpacks you need be root to perform this command
```

위의 에러는 관리자(root) 계정이 아닌 사용자 계정으로 yum을 실행하려고 해서 발생<br> 해결 방법은 명령창에 아래와 같이 입력하여 관리자(root) 모드로 전환

```
# su -<br>
암호: {관리자 비밀번호 입력}
```


### 2.인터넷 연결 문제로 인한 에러

vim 설치하기 위해 yum 명령어를 사용하여 인스톨을 실행시 아래의 에러가 발생

```
loaded plugins: fastestmirror
```

위의 에러는 인터넷이 연결되어 있지 않아서 발생하는 경우가 많음<br>
아래와 같이 입력해서 인터넷이 설정 되어 있는지 확인

```
vi /etc/sysconfig/network-scripts/ifcfg-enp0s3
```

위의 명령어를 실행하면 아래와 같은 창이 생성


CentOS인터넷1.PNG<br>
ONBOOT가 no 설정되어 인터넷 연결이 안됨<br>
아래와 같이 설정을 수정

CentOS인터넷2.PNG<br>
변경 하고자 하는 곳으로 이동 한뒤 insert키 ```i```를 눌러 yes로 수정<br>
수정을 한뒤 esc를 누르고 ```:wq```를 입력하고 나오자.

변경한 값이 설정 되도록 아래와 같이 입력하여 재부팅

```
# reboot
```

인터넷이 원활하게 연결되는지 확인하기 위해 테스트를 진행 (8.8.8.8은 구글의 주소 값)

```
# ping 8.8.8.8 
```

원활하게 연결 된다면 패킷을 내보 내는 것을 확인 할 수 있음<br>
종료를 하기 위해서는 Ctrl + C를 클릭하여 종료
